```python
# Import modules and libraries

from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split


# Generate dataset using make_blobs with 150 samples and 3 classes 
centers = [[2, 4], [6, 6], [1, 9]]
data, labels = make_blobs(n_samples=150, centers=np.array(centers), random_state=1)


# Data splitting : 80% (Training dataset) and 20% (Test dataset)
train_data, test_data, train_labels, test_labels = train_test_split(data, labels, train_size=0.8, test_size=0.2, random_state=12)


# Perform KNN analysis by importing the KNeighborsClassifier
# Initialize using minkowski distance (p=2 for Euclidean distance) and 5 neighbors
knn = KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2)

# Train the KNN classifier on the training dataset to learn the patterns
knn.fit(train_data, train_labels)

# Predict on test data
# Use the trained KNN model to predict labels for the test dataset.
test_data_predicted = knn.predict(test_data)

# Calculate and print accuracy score by comparing predicted values(test_data_predicted) with actual test labels(test_labels)
# this score will tell us how well the model has perform on the test dataset
accuracy = accuracy_score(test_data_predicted, test_labels)

#Print the accuracy score
print(f"Accuracy Score of KNN on test dataset: {accuracy}")

# Plot the results
# Visualize the dataset with their true labels with a figure size of 10x6 inches
plt.figure(figsize=(10, 6))


# Scatter plot of the entire dataset with true labels
#  X-axis is feature 1 values i.e. data[:, 0]
#  Y-axis is feature 2 values i.e. data[:, 1]
#  c=labels: Color the points based on their true labels
#  cmap='viridis': Using the default 'viridis' colormap
#  marker='o': marker shapes to be circle
#  edgecolor='k': Add black boundary to the markers
#  s=50: Setting marker size to 50
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis', marker='o', edgecolor='k', s=50)
plt.title('Data Points with True Labels')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()

# Scatter plot of the test data with predicted labels
#  X-axis is feature 1 values i.e. test_data[:, 0]
#  Y-axis is feature 2 values i.e. test_data[:, 1]
plt.figure(figsize=(10, 6))
plt.scatter(test_data[:, 0], test_data[:, 1], c=test_data_predicted, cmap='viridis', marker='s', edgecolor='k', s=50)
plt.title('KNN classification on test data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
